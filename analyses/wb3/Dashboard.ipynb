```python
#!source ~/.bashrc
#!source_wholebrain2.0 
#!pwd
import os, glob 

analysis_dir="/data/NIMH_scratch/kleinrl/analyses/wb3/"

projects = glob.glob(analysis_dir+"/*/")

#projects=[f for f in analysis_dir.iterdir() if f.is_dir()]

#print(projects)

for p in projects: 
    seeds = glob.glob(p+"/fsl_feat*")
    p_name = [ pp for pp in p.split('/') if pp != '' ][-1]

    #print("    "+p)
    #print(seeds)
    for s in seeds: 
        ds              = glob.glob(s+"/DAY0/ds*")
        ds_len          = len(ds)

        zstats          = glob.glob(s+"/DAY0/ds*/ds*.feat/thresh_zstat1.nii.gz")
        zstats_len      = len(zstats)

        smoothed        = glob.glob(s+"/mean/thresh_zstat1.fwhm*.nii.gz")
        smoothed_len    = len(smoothed)


        print("{}    {}    {}    {}    {}"
                .format(
                    p_name.ljust(30), 
                    s.split('/')[-1].ljust(30), 
                    str(ds_len).ljust(5), 
                    str(zstats_len).ljust(5), 
                    str(smoothed_len).ljust(5)))

```
